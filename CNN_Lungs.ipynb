```python
import os
import csv
```

```python
#Import all Normal Images into one CSV file
```

```python
# Define the directory containing your dataset
dataset_dir = './Normal/images/'
```

```python
# Define the CSV file to save the metadata
output_csv = "normal_image_data.csv"
```

```python
# Initialize an empty list to store data
image_data = []
```

```python
# Traverse the dataset directory
for label in os.listdir(dataset_dir):
    label_dir = os.path.join(dataset_dir, label)
    print(label)
    print(label_dir)
    image_data.append([label_dir, label])
```

```python
print(image_data)
```

```python
# Save the image data to a CSV file
with open(output_csv, mode='w', newline='') as file:
    writer = csv.writer(file)
    # Write the header
    writer.writerow(["image_path", "label"])
    # Write the image data
    writer.writerows(image_data)
```

```python
# Display the number of images
print(f"Total images: {len(image_data)}")
```
